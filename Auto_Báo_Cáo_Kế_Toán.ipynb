*   B3: Trong Sheet "SB", xóa hết đơn vị tiền tệ ở cột "Spend" và "Sale". Ví dụ "Spend (CAD)" thì thành "Spend"

```python
import pandas as pd
import numpy as np
import re
from google.colab import files
import io
```

```python
uploaded = files.upload()
```

```python
dfs = {}

for file_name in uploaded.keys():
    xls = pd.ExcelFile(file_name)
    sheet_names = xls.sheet_names

    for sheet_name in sheet_names:
        dfs[sheet_name] = pd.read_excel(file_name, sheet_name=sheet_name)

SP = dfs['SP']
SD = dfs['SD']
SB = dfs['SB']
```

```python
SP_data = SP.groupby('Advertised SKU').agg({'Spend':'sum', '7 Day Total Sales ':'sum'})
SP_data.rename(columns={'Advertised SKU':'SKU', "Spend":"Spend", "7 Day Total Sales ":"Sales"},inplace = True)
SP_data.reset_index()
```

```python
SD_data = SD.groupby('Advertised SKU').agg({'Spend':'sum', '14 Day Total Sales ':'sum'})
SD_data.rename(columns={'Advertised SKU':'SKU', "Spend":"Spend", "14 Day Total Sales ":"Sales"},inplace = True)
SD_data.reset_index()
```

```python
def extract_keywords(campaign):
    pattern = r'(\S+)(?=\s+(?:Video|SB))'
    matches = re.findall(pattern, campaign)
    return ', '.join(matches)

if 'SB' in locals():
    SB['Advertised SKU'] = SB['Campaigns'].apply(extract_keywords)
```

```python
SB_data = SB.groupby('Advertised SKU').agg({'Spend':'sum', 'Sales':'sum'})
SB_data.rename(columns={'Advertised SKU':'SKU', "Spend":"Spend", "Sales":"Sales"},inplace = True)
SB_data.reset_index()
```

```python
data = pd.concat([SP_data, SD_data, SB_data])
data.reset_index()
```

```python
final_data = data.groupby('Advertised SKU').agg({'Spend':'sum', 'Sales':'sum'})
final_data = final_data.astype({'Spend':'float', 'Sales':'float'})
final_data.reset_index()
```

```python
final_data.to_excel('Dữ liệu cho kế toán.xlsx')
files.download("Dữ liệu cho kế toán.xlsx")
```
